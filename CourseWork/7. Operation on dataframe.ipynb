```python
import pandas as pd
# reading the csv file
df_salary=pd.read_csv("Salary.csv")
df_salary.head()
```

# Maths Operations

## 1.1 Operations in Columns

```python
# select a column and calculate total sum 
import numpy as np

# assuming df_salary already exists
age = np.random.randint(18, 60, 14017)
df_salary = df_salary.copy()  # optional
df_salary['AGE'] = age        # or pd.Series(age, index=df_salary.index)

total_age = df_salary['AGE'].sum()
print(total_age)
```

```python
# count , mean . std, max and min 
df_salary['AGE'].count()
```

```python
df_salary['AGE'].max()
```

```python
df_salary['AGE'].min()
```

```python
df_salary['AGE'].mean()
```

```python
df_salary['AGE'].std()
```

```python
body_count = np.random.randint(1, 1000, 14017)
df_salary = df_salary.copy()  # optional
df_salary['Body Count'] = body_count        # or pd.Series(age, index=df_salary.index)
df_salary
```

```python
# easier calculation with .discribe()
df_salary[['AGE', 'Body Count']].describe()
```

```python
# calculating the sum in a row 
df_salary['Rand Score']=(df_salary['AGE'] + df_salary['Body Count'])
```

```python
import pandas as pd
# reading the csv file
df_salary=pd.read_csv("Salary.csv")
df_salary.head()
```

# 1 Value Counts

```python
import numpy as np
# create random numbers for our new score columns
tenth_score   = np.random.randint(35, 100, size=14017)
twelfth_score = np.random.randint(35, 100, size=14017)

# create a series using the random numbers
series1 = pd.Series(tenth_score,   index=np.arange(0, 14017))
series2 = pd.Series(twelfth_score, index=np.arange(0, 14017))
series1 = pd.Series(tenth_score)
series2 = pd.Series(twelfth_score)
# using assign() to add multiple columns
df_salary = df_salary.assign(Tenth_score=series1, twelfth_score=series2)
```

```python
df_salary
```

```python
# counting gender elements
# len function 
len(df_salary['Tenth_score'])
# .count() method
df_salary['Tenth_score'].count()
```

```python
# counting gender elements by category
df_salary['Tenth_score'].value_counts()
```

```python
# return the relative freq ( divide all values by the sum of values )
df_salary['Tenth_score'].value_counts(normalize=True)
```

```python
#  counting " parental level of education" elements by category
df_salary['JobTitle'].value_counts()
```

```python
# return the relative frequency and round to 2 decimals 
df_salary['JobTitle'].value_counts(normalize=True)
```
